```python
import numpy as np
from scipy.misc import imread
image_id = '0271c9e4-bbab-11e8-b2ba-ac1f6b6435d0'
image = np.zeros(shape=(512,512,4))
image[:,:,0] = imread(train_path + image_id + "_red" + ".png")
image[:,:,1] = imread(train_path + image_id + "_blue" + ".png")
image[:,:,2] = imread(train_path + image_id + "_green" + ".png")
image[:,:,3] = imread(train_path + image_id + "_yellow" + ".png")
```

```python
fig, ax = plt.subplots(1,5,figsize=(30,100))
ax[1].set_title('stained microtubules (red)')
ax[1].imshow(image[:,:,0].astype(np.int), cmap="Reds")
ax[2].set_title('stained nucleus (blue)')
ax[2].imshow(image[:,:,1].astype(np.int), cmap="Blues")
ax[0].set_title('Labels: 6')
ax[0].imshow(image[:,:,2].astype(np.int), cmap="Greens")
ax[3].set_title('stained endoplasmatic reticulum (yellow)')
ax[3].imshow(image[:,:,3].astype(np.int), cmap="Oranges")
ax[4].set_title('RGB Image')
ax[4].imshow(image[:,:,0:3].astype(np.int))
```
